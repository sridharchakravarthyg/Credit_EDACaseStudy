##### Defining the function to retrieve metadata and finding the metadata of the prev_application dataframe stored in data_merge_2

```python
def meta(x):
    Nulls = x.isnull().sum()
    Null_Percent = (x.isnull().sum()/len(x))*100
    Unique = x.nunique()
    Datatypes = x.dtypes
    return pd.concat([Datatypes, Unique,Nulls, Null_Percent], axis=1, keys=['Data_Type', 'Unique','Nulls', 'Null_Percent' ])

data_2_info=meta(data_merge_2)
data_2_info.reset_index(drop=False)
```

##### Finding the null value percentage of columns having minimum null value of percentage of 40 (i.e cutoff percentage)

```python
plt.figure(figsize = (15,5))
filtered_cols=data_2_info[(data_2_info.Null_Percent>40)]
filtered_cols.Null_Percent.plot.bar()
```

##### Creating new dataframe merge_loan to perform analysis

```python
merge_loan = pd.merge(left=data_merge_1 , right = data_merge_2 , how='inner' , on='SK_ID_CURR')
merge_loan.shape
```

##### Creating a new coloumn Default_History in the dataframe merge_loan based on the 'TARGET' Variable

```python
merge_loan['Default_History'] = merge_loan['TARGET']
merge_loan['Default_History'].replace({0: "Non-Defaulter", 1: "Defaulter"}, inplace=True)
merge_loan.head()
```

```python
merge_loan.nunique().sort_values()
```

#Creating dataframes to classify the TARGET values based on Defaulters and non-Defaulters ; 
- 0 for Non-Defaulters
- 1 for Defaulters

Also created a new dataframe filter_loan to only select a few relevant coloumns to perform analysis on

```python
merge_loan_1 = merge_loan[merge_loan['TARGET']==1]
merge_loan_0 = merge_loan[merge_loan['TARGET']==0]
filter_loan = merge_loan.loc[:,['TARGET','AMT_DOWN_PAYMENT', 'RATE_DOWN_PAYMENT', 'DAYS_TERMINATION', 'DAYS_LAST_DUE', 'DAYS_LAST_DUE_1ST_VERSION', 'DAYS_FIRST_DUE', 'DAYS_FIRST_DRAWING', 'CNT_PAYMENT', 'DAYS_DECISION', 'AMT_APPLICATION']]
filter_loan_0 = filter_loan[filter_loan['TARGET']==0]
filter_loan_1 = filter_loan[filter_loan['TARGET']==1]
```

### Creating a heatmap to analyse the correlation between the coloumns

```python
checkvar=filter_loan.select_dtypes(include=['object']).columns
filter_loan_corr=pd.get_dummies(filter_loan, columns=checkvar,drop_first=True)

plt.figure(figsize=(25,20))

sns.heatmap(filter_loan_corr.corr(), annot = True,cmap='coolwarm')
plt.show()
```
